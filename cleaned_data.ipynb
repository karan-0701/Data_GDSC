```python
# import libraries
import numpy as np
import pandas as pd

# for data visualisation
import matplotlib.pyplot as plt
```

```python
# import data
data = pd.read_csv("/Users/karandeepsingh/Desktop/50_Startups.csv")
data
```

```python
# no conventional na values
data.isna().sum()
```

```python
# zero values exist so replacing them with mean
mean_value = np.mean(data['R&D Spend'])
mean_value

data['R&D Spend'] = np.where(data['R&D Spend'] == 0, mean_value, data['R&D Spend'])
data['Administration'] = np.where(data['Administration'] == 0, np.mean(data['Administration']), data['Administration'])
data['Marketing Spend'] = np.where(data['Marketing Spend'] == 0, np.mean(data['Marketing Spend']), data['Marketing Spend'])
data['Profit'] = np.where(data['Profit'] == 0, np.mean(data['Profit']), data['Profit'])
```

```python
# finiding outliers uisng plots
plt.scatter(data['R&D Spend'], data['Profit'])
plt.scatter(data['Marketing Spend'], data['Profit'])

# No major outilers found
```

```python
# making dummy variables for the different state

dataf = pd.get_dummies(data['State'])
data = pd.concat([data, dataf], axis = 1)
data.drop(columns=['State'])
```
